# Constuction bdd

```python
import pandas as pd
```

## BDD 2020

```python
enedis_2020_1=pd.read_csv("Données/enedis_api_1.csv")
enedis_2020_2=pd.read_csv("Données/enedis_api_2.csv")
enedis_2020_3=pd.read_csv("Données/enedis_api_3.csv")
enedis_2020_4=pd.read_csv("Données/enedis_api_4.csv")
enedis_2020_5=pd.read_csv("Données/enedis_api_5.csv")
```

```python
enedis_2020=pd.concat([enedis_2020_1,enedis_2020_2,enedis_2020_3,enedis_2020_4,enedis_2020_5])
enedis_2020.drop_duplicates(inplace=True)
enedis_2020.sort_values("horodate",inplace=True)
```

```python
enedis_2020.to_csv("Données/enedis_2020.csv",index=False)
```

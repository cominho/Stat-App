# Construction bdd finale merged avec données température

```python
import pandas as pd 
```

## Données ENEDIS

"Ce jeu de données restitue la consommation d’électricité au pas 1/2h des points de soutirage <= 36kVA raccordés au réseau Enedis. Il donne les volumes d'énergie soutirés, les courbes de charge moyennes de clients dotés de compteurs communicants et le nombre de clients. Ces agrégats sont disponibles par plage de puissance souscrite, profil et région."

Dictionnaire des variables : https://data.enedis.fr/api/datasets/1.0/conso-inf36-region/attachments/note_externe_jdd_article_23_pdf/

```python
enedis_2020=pd.read_csv("Données/enedis_2020.csv")
enedis_2021=pd.read_csv("Données/enedis_2021.csv")

enedis=pd.concat([enedis_2020,enedis_2021])
enedis.drop_duplicates(inplace=True)
enedis.sort_values("horodate",inplace=True)
```

```python
enedis["profil_type"]=enedis["profil"].apply(lambda x : x[0:3])
enedis
```

```python
enedis.drop(enedis[enedis["profil_type"]!="RES"].index,inplace=True)
```

On récupère la date :


```python
## On extrait la date pour classifier par jour 
enedis["date"]=enedis["horodate"].apply(lambda x : x[0:10])
enedis
```

### Données sur les nombres de ménage : 

```python
menage=pd.read_csv("Données/nb_ménages.csv",sep=";")
menage
```

```python
enedis_menage=enedis.merge(menage,how="left",right_on="code_reg",left_on="code_region")
```

```python
enedis_menage
```

Ajout de la conso moyenne (conso totale / nb de ménages ):

```python
enedis_menage["conso_moyenne"]=enedis_menage["total_energie_soutiree_wh"]/enedis_menage["Ensemble"]
```

## Données méteo


**Données des températures régionales du 1er janvier 2018 au 31 octobre 2022 :**

```python
meteo_reg_df=pd.read_csv("Données/temperature-quotidienne-regionale.csv",";")
meteo_reg_df
```

```python
meteo_3h=pd.read_csv("Données/donnees-temperatures-3h.csv",sep=";")
```

```python
meteo_3h_reduce=meteo_3h[['Date',"Température (°C)",'communes (name)','region (name)','region (code)']]
meteo_3h_reduce.dropna(inplace=True)
meteo_3h_reduce
```

On calcule la température moyenne par région pour toutes les trois heures :

```python
meteo_3h_reduce["region (code)"]=meteo_3h_reduce["region (code)"].astype("int")
meteo_3h_reduce["region (code)"]=meteo_3h_reduce["region (code)"].astype("str")
```

```python
meteo_3h_mean=meteo_3h_reduce.groupby(["Date","region (code)"])["Température (°C)"].mean().reset_index()
```

```python
meteo_3h_mean
```

## Données socio économique

```python

socio=pd.read_csv("Données/data_socio.csv",sep=";",header=2)
socio
```

```python
socio["Code"]=socio["Code"].astype("str")
socio.drop( socio[ (socio['Code'] == "1") | (socio['Code'] == "2") | (socio['Code'] == "3") | (socio['Code'] == "4") | (socio['Code'] == "6")].index, inplace=True)
socio
```

## Base de données merged

```python
enedis_menage["code_region"]=enedis_menage["code_region"].astype("str")
```

```python
enedis_socio=pd.merge(enedis_menage,socio,how='left',left_on="code_region",right_on="Code")
```

Finalement, il faut merger les données météo et enedis sur date et région

```python
meteo_reg_df["code_insee_region"]=meteo_reg_df["code_insee_region"].astype("str")
```

```python
enedis_final=pd.merge(enedis_socio,meteo_reg_df,how='left',left_on=["date","code_region"],right_on=["date","code_insee_region"])
```

```python
enedis_final.drop(columns=["region_y","id","tmin","tmax","code_insee_region"],inplace=True)
```

```python
enedis_final.to_csv("Données/enedis_final.csv",index=False)
```

```python
enedis_final
```

### Table enedis pour toutes les 3H :

```python
meteo_3h_mean["date_cut"]=meteo_3h_mean["Date"].str[0:19]
enedis_final["date_cut"]=enedis_final["horodate"].str[0:19]
enedis_final["code_region"]=enedis_final["code_region"].astype("str")
```

```python
enedis_final_3h=pd.merge(enedis_final,meteo_3h_mean,how='left',left_on=["date_cut","code_region"],right_on=["date_cut","region (code)"])
```

```python
enedis_final_3h.dropna(inplace=True)
```
